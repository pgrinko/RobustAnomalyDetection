```python
from robustad import RobustAnomalyDetector
#Import required libraries here
```

```python
data = pd.read_csv('original_data.csv')['data_values'].values
#Converting to numpy array to avoid failures
```

```python
%pylab inline
pd.Series(data).plot(figsize = (15, 6), title = 'Original data', linewidth = 3)
#Plotting the original data. It seems to have evident seasonality, but we don't mind this and want to try find it automatically
```

```python
model = RobustAnomalyDetector(alpha = 8)
model.fit(data)
#There are no problems with data fitting, the warning cautions us that there is the real reason to remove seasonality and so it does
```

```python
result = model.predict(anomaly_kind = 'peak')
#We want to see anomalies that are over acceptable level
```

```python
pd.Series(data).plot(figsize=(15, 6), title='Original data with obtained anomalies', linewidth = 3)
plt.scatter(result, data[result], color = 'r', s = 600)
#That is it, we see highlighted circle areas that suppose to be over acceptable data values according to given variable makes through the time
```
